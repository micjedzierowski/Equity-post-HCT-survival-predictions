Defining functions for plotting histograms.

```python
def plot_histograms(df, rows, cols):
  rows = rows
  cols = cols
  fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*5))
  axes = axes.flatten()
  for i, col in enumerate(df.columns):
      sns.histplot(df, x=col, ax=axes[i], kde=True, color='blue')
      axes[i].set_title(f'Distribution of {col}')
      axes[i].set_xlabel(col)
      axes[i].set_ylabel('Frequency')


  plt.tight_layout(pad=5.0)
  plt.show()

def plot_histogram(df, var):
  sns.histplot(data=df[[var]], x=var, kde=True, color='blue')
  plt.title(f'Distribution of {var}')
  plt.xlabel(f"{var}")
  plt.ylabel('Frequency')
  plt.show()
```

Based on histograms and correlation matrix for HLA variables I have decided to create a ratio between hla_high and hla_low corresponding values as they are discrete. As some of values in denominator were zeros, in order to avoid going to infinity for combined_param, I have replaced output of calculated combined_param with max value for numerator.

```python
def combined_hla_high_low(df, variables):
    new_hla_params = []
    for col in variables:
      if 'high' in col:
          low_col = col.replace('high', 'low')
          if low_col in df.columns:
              high = df[col].fillna(0)
              low = df[low_col].fillna(0)

              combined_param = np.where(
                  (high == 0) & (low == 0),
                  0,
                  np.where(low == 0, high.max(), (high / low).round(2))
              )
              combined_param = pd.Series(combined_param, index=df.index, name=f"{col}_low_ratio")
              new_hla_params.append(combined_param)

    return new_hla_params

combined_params = combined_hla_high_low(df=x_train, variables=numerical_hla_variables)
combined_params = pd.DataFrame(combined_params).T
```

```python
from scipy.stats import ks_1samp, norm

# Standardize age_at_hct to compare with standard normal distribution
age_standardized = (x_train['age_at_hct'] - x_train['age_at_hct'].mean()) / x_train['age_at_hct'].std()

# Perform KS test against normal distribution
stat, p = ks_1samp(age_standardized, norm.cdf)

print(f"KS Statistic: {stat.round(2)}")
print(f"P-value: {p.round(2)}")

# Interpretation
if p < 0.05:
    print("Reject H0: Data does NOT follow a normal distribution.")
else:
    print("Fail to reject H0: Data appears to be normally distributed.")
```

```python
Q1 = x_train['age_at_hct'].quantile(0.25)  # 25th percentile
Q3 = x_train['age_at_hct'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile range

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = x_train[(x_train['age_at_hct'] < lower_bound) | (x_train['age_at_hct'] > upper_bound)]

print(f"Number of Outliers: {outliers.shape[0]}")
print(outliers[['age_at_hct']])
```

```python
from scipy.stats import zscore

x_train['age_zscore'] = zscore(x_train['age_at_hct'])
outliers_z = x_train[np.abs(x_train['age_zscore']) > 3]  # 3 standard deviations

print(f"Number of Outliers: {outliers_z.shape[0]}")
print(outliers_z[['age_at_hct']])
```

For categorical variables, especially from medical field, it is necessary to conduct statistical test of significance. Below is an example of chi-square test for two given variables: donor_related and melphalan_dose. First of all we need to create a table of contingency that shows how often values of both variables exist in our dataset.

**Chi-square test tells us if there is significant association between variables however we don't know how strong or weak it is.**

```python
contingency_table = pd.crosstab(df_cleaned_categorical['donor_related'], df_cleaned_categorical['melphalan_dose'])
print(contingency_table)
```

In general for this pair and all pairs of variables we are going to make hypotheses as following:

Null Hypothesis (H_0) - there is no association between two variables, they are independent on each other,

Alternative Hypothesis (H_1) -  there is association between two variables, they are dependent on each other.

For alpha a = 0.05 and P-value p, where p <= a, we conclude that there is a statistically significant association between donor_related and melphalan_dose.

In the other hand if p > a, we conclude that there is no evidence of a significant association between donor_related and melphalan_dose.

```python
# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2 Statistic: {chi2.round(2)}")
print(f"P-value: {p.round(2)}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected.round(2))
```

From the chi-square test for donor_related and melphalan_dose we can observe that p ~ 0.0 which is below a = 0.05 treshold that tells us **there is significant association** between two of them.

**Cramér's V measures the strength of association between two categorical variables. It focuses on effect's size.**

Results close to 0 indicate there is a weak or negligible association between variables, from the other hand results closer to 1 indicate there is strong association.

```python
def cramers_v(df, variable_1, variable_2):
    contingency_table = pd.crosstab(df[variable_1], df[variable_2])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    N = df_cleaned_categorical.shape[0]
    k = min(contingency_table.shape)
    return np.sqrt(chi2 / (N * (k - 1)))
```

Calculating Cramer's V matrix for categorical variables in df_cleaned_categorical

```python
def calculate_cramers_v_matrix(df, categorical_vars):
    results = pd.DataFrame(index=categorical_vars, columns=categorical_vars)
    for var1 in categorical_vars:
        for var2 in categorical_vars:
            if var1 == var2:
                results.loc[var1, var2] = 1.0  # Diagonal is always 1 (perfect correlation)
            else:
                results.loc[var1, var2] = cramers_v(df=df, variable_1=var1, variable_2=var2)

    return results.astype(float).round(3)

cramers_v_matrix = calculate_cramers_v_matrix(df=df_cleaned_categorical, categorical_vars=df_cleaned_categorical.columns)
```

Steps to Impute Missing Values with KNN
1. Preprocess Data:

* Separate the rows where donor_age is missing from the rest of the dataset.
* Ensure all categorical variables are encoded (e.g., using one-hot encoding or ordinal encoding).
* Normalize numerical features to ensure they are on a similar scale.
2. Train KNN Model:

* Use rows with known donor_age values to train the KNN model.
* Select an appropriate number of neighbors (k).
3. Predict Missing Values:

* Use the trained model to predict the missing values for donor_age.

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

def impute_donor_age(df, target_column='donor_age'):
    # Separate features and target
    features = df.drop(columns=[target_column])
    target = df[target_column]
    # Identify numerical and categorical columns
    numerical_cols = list(combined_params.columns) + numerical_non_hla_variables
    numerical_cols.remove('year_hct')
    numerical_cols.remove('donor_age' )
    categorical_cols = list(set(features.columns) - set(numerical_cols))

    def frequency_encoder(df, cols=categorical_cols):
      for col in cols:
          freq_map = df[col].value_counts().to_dict()
          df[col] = df[col].map(freq_map)
      return df

    # Encode categorical variables and scale numerical variables
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_cols),
            ('cat', FunctionTransformer(frequency_encoder), categorical_cols)  # Encoding can be added here if necessary
        ],
        remainder='passthrough'
    )

    # Create a new DataFrame for imputation
    data_for_imputation = df.copy()
    # Create KNN imputer pipeline
    knn_imputer = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('imputer', KNNImputer(n_neighbors=5))
    ])
    # Fit KNN imputer on the entire dataset
    imputed_data = knn_imputer.fit_transform(data_for_imputation)

    # Convert back to DataFrame
    imputed_df = pd.DataFrame(imputed_data, columns=df.columns, index=df.index)
    # Return the DataFrame with imputed values
    return imputed_df

# Example usage
imputed_knn_df = impute_donor_age(df=df_cleaned, target_column='donor_age')
```

# **Survival function and cumulative hazard function**
The survival function shows how many patients survive for how long (Wikipedia: Kaplan–Meier estimator). At month 0, 100 % of the patients live. At month 20, only 40 % – 60 % remain, depending on their race group. Patients with "more than one race" have the highest probability of survival, whites the lowest.

For those who are used to working with cumulative density functions (cdf) of probability distributions, the survival function is nothing else than a top–down mirrored cdf of the time-of-event probability distribution.

The cumulative hazard is another representation of the same facts; it corresponds to the negative logarithm of the survival function (Wikipedia: Nelson–Aalen estimator).

### HLA Matching Variables (e.g., hla_match_a_high, hla_match_b_low, hla_match_dqb1_low, etc.)
HLA matching is critical for transplantation success. These variables likely represent different levels of matching for specific HLA loci. The values indicate whether a certain HLA match is "high", "low", or missing (indicating no match).

Mutual Exclusivity:
For each HLA locus (e.g., A, B, C, DRB1, DQB1), a donor and recipient can only have one type of match (high, low, or none).
For example, hla_match_a_high and hla_match_a_low are likely mutually exclusive, meaning both can't be true at the same time.
Similarly, hla_match_b_high and hla_match_b_low are mutually exclusive.
Interpretation:
HLA Match Level: These variables suggest whether the donor and recipient have high or low compatibility at each locus, which impacts graft-versus-host disease (GVHD) and transplant success.
A high match (e.g., hla_match_a_high) is usually associated with better outcomes, while low match (e.g., hla_match_b_low) increases risks of complications.

### HLA Resolution Variables (e.g., hla_high_res_8, hla_low_res_6, hla_high_res_6)
These variables likely indicate the resolution level of HLA typing, which refers to the level of detail in matching HLA alleles. Higher resolution means more precise matching, often at the allele level, while lower resolution indicates matching at the group level.

Mutual Exclusivity:
A given locus (e.g., hla_high_res_8, hla_low_res_6) cannot have multiple resolutions. You would either have high resolution at a particular locus or low resolution, but not both.
Interpretation:
High Resolution Typing: More precise matching at the allele level, improving the chances of transplant success.
Low Resolution Typing: Less precise and may increase the risk of mismatches and complications.

### HLA Matching Variables for Different Loci (e.g., hla_match_a_high, hla_match_b_low, hla_match_c_high)
These represent matches for different HLA loci: A, B, C, DRB1, DQB1, etc. HLA matching is performed across multiple loci to increase transplant success.

Mutual Exclusivity:
Different loci are independent, but for each locus, a single match category (high, low) should be chosen.
Interpretation:
Different HLA loci may have different levels of importance in transplant success. A mismatch at one locus (e.g., hla_match_a_low) may be acceptable if there is a perfect match at another locus (e.g., hla_match_b_high).
Example:
A patient may have high matches at HLA-A and HLA-B but low matches at HLA-DRB1 and HLA-DQB1, which can influence GVHD and overall survival rates.

### Demographic Variables (e.g., donor_age, age_at_hct)
These variables represent key patient and donor demographic information, such as age at the time of transplant and donor age.

Mutual Exclusivity:
These variables are not mutually exclusive in themselves, but they may be related. For example, donor age (donor_age) and recipient age (age_at_hct) could interact, where a younger donor may be preferred, but the recipient’s age also affects graft survival.
Interpretation:
Donor Age: Younger donors are typically preferred because they tend to have better cell viability and lower rates of complications.
Recipient Age: Older recipients may have higher risks of complications and poorer transplant outcomes.

### Comorbidity and Clinical Scores (e.g., comorbidity_score, karnofsky_score)
These are clinical scores that assess comorbidities and performance status, which are important in determining transplant eligibility and predicting outcomes.

Mutual Exclusivity:

These variables aren’t directly mutually exclusive, but they interact. For example, a higher comorbidity score might suggest a lower Karnofsky performance score (indicating poorer overall health and function).
Interpretation:

Comorbidity Score: Higher scores indicate more health problems and may increase the risk of transplant failure.
Karnofsky Score: This score reflects the patient’s ability to perform daily activities, with lower scores indicating poorer health status, which can affect transplant success.

### Year of HCT (year_hct)
This variable represents the year the transplant occurred and might have an impact on outcomes due to advancements in transplant techniques over time.

Mutual Exclusivity:
This variable is not mutually exclusive with others but is important for temporal trend analysis.
Interpretation:
Transplants performed in more recent years may benefit from improved protocols, better post-transplant care, and newer immunosuppressive drugs, leading to better outcomes.
